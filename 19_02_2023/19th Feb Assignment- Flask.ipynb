#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python. It is called a "micro" framework because it does not require particular tools or libraries, and is designed to keep the core of the application simple and scalable. Flask is known for its simplicity, flexibility, and ease of use, and is popular among developers for building small to medium-sized web applications and APIs.

Some advantages of using Flask include:

* Lightweight and simple: Flask is a lightweight and simple framework that can be easily understood and learned by developers, and it doesn't require a lot of boilerplate code to get started.

* Easy to use and configure: Flask is easy to set up and configure, with a simple and intuitive API that makes it easy to build web applications quickly.

* Highly customizable: Flask is highly customizable, allowing developers to add or remove features as needed, and easily integrate it with other libraries or tools.

* Large community and ecosystem: Flask has a large community of developers, which means there are many plugins, extensions, and libraries available to extend its functionality.

* Flexible templating engine: Flask comes with a flexible templating engine called Jinja2, which allows developers to easily create dynamic HTML pages.

* Integrated development server: Flask comes with an integrated development server, which allows developers to easily test their applications during development.

#### Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello, World!!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

#### Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to specific Python functions, also known as "view functions". This is done using the @app.route decorator, which is used to define the URL pattern for a view function.

App routing is important in Flask because it allows developers to define the URLs and behavior of their web application in a simple and intuitive way. By mapping URLs to view functions, developers can define the structure and behavior of their web application, and handle different types of HTTP requests and responses.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL for the specified endpoint, based on the URL rules defined in the application.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}'

with app.test_request_context():
    # generate a URL for the index() function
    print(url_for('index'))

    # generate a URL for the profile() function with username = 'Prince'
    print(url_for('profile', username='Prince'))
```
